## Import Semua Packages/Library yang Digunakan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

Data items

```python
items_df = pd.read_csv("https://raw.githubusercontent.com/stefanawndprst/e-commerce-public/main/order_items_dataset.csv")
items_df.head()
```

data payment

```python
payment_df = pd.read_csv("https://raw.githubusercontent.com/stefanawndprst/e-commerce-public/main/order_payments_dataset.csv")
payment_df.head()
```

data translate

```python
translate_df = pd.read_csv("https://raw.githubusercontent.com/stefanawndprst/e-commerce-public/main/product_category_name_translation.csv")
translate_df.head()
```

data mengenai product

```python
prdct_df = pd.read_csv("https://raw.githubusercontent.com/stefanawndprst/e-commerce-public/main/products_dataset.csv")
prdct_df.head()
```

mengambil 2 kolom dari data product

```python
prdct1_df = prdct_df[["product_id","product_category_name"]]
prdct1_df.head()
```

menggabungkan data product dengan translate

```python
product_df = pd.merge(
    left= translate_df,
    right= prdct1_df,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)
product_df.head()
```

menggabungkan data item dengan payment

```python
order_df = pd.merge(
    left=items_df,
    right=payment_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
order_df.head()
```

menggabungkan gabungan dua data yang telah digabung

```python
order_product_df = pd.merge(
    left= order_df,
    right= product_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
order_product_df.head()
```

### Assessing Data

```python
order_product_df.info()
```

```python
time_feature = ["shipping_limit_date"]

for i in order_product_df.columns:
  for j in time_feature:
    if i == j:
      order_product_df[i]= pd.to_datetime(order_product_df[j])
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
payments_df = order_product_df.groupby(by="payment_type").order_id.nunique().reset_index()
payments_df.rename(columns={
    "order_id": "payment_count"
}, inplace=True)
payments_df
plt.figure(figsize=(10, 5))
colors_ = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
sns.barplot(
    x="payment_count",
    y="payment_type",
    data=payments_df.sort_values(by="payment_count", ascending=False),
    palette=colors_
)
plt.title("Number of payment", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
```

### Pertanyaan 2:

```python
products_df = order_product_df.groupby(by="product_category_name_english").product_id.nunique().reset_index()
products_df.rename(columns={
    "product_id": "product_count"
}, inplace=True)
products_df
plt.figure(figsize=(25, 15))
colors_ = ["#72BCD4"]
sns.barplot(
    x="product_count",
    y="product_category_name_english",
    data=products_df.sort_values(by="product_count", ascending=False),
    palette=colors_
)
plt.title("Number of Product", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
```
